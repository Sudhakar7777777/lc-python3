# Solving equations with Scipy Linear Algebra
# 2w + x + 3y + 5z = 10
# 2.3w + 2x + 5y + z = 8
# -3.2w + 2x + 3y + 8z = 3
# -0.9w + 3x - 5y - 8z = -20

```python
#importing the scipy and numpy packages
from scipy import linalg
import numpy as np

#Declaring the numpy arrays
a = np.array([[2, 1, 3, 5], 
              [2.3, 2, 5, 1], 
              [-3.2, 2, 3, 8],
              [-0.9, 3, -5, -8]])

b = np.array([10, 8, 3, -20])

#Passing the values to the solve function
x = linalg.solve(a, b)

#printing the result array
print(x)
```

```python
import scipy as scipy
```

```python
xx = [ 3.08,  3.1 ,  3.12,  3.14,  3.16,  3.18,  3.2 ,  3.22,  3.24,
    3.26,  3.28,  3.3 ,  3.32,  3.34,  3.36,  3.38,  3.4 ,  3.42,
    3.44,  3.46,  3.48,  3.5 ,  3.52,  3.54,  3.56,  3.58,  3.6 ,
    3.62,  3.64,  3.66,  3.68]
```

```python
yy = [ 0.000857,  0.001182,  0.001619,  0.002113,  0.002702,  0.003351,
    0.004062,  0.004754,  0.00546 ,  0.006183,  0.006816,  0.007362,
    0.007844,  0.008207,  0.008474,  0.008541,  0.008539,  0.008445,
    0.008251,  0.007974,  0.007608,  0.007193,  0.006752,  0.006269,
    0.005799,  0.005302,  0.004822,  0.004339,  0.00391 ,  0.003481,
    0.003095]
```

```python
print(len(xx))
print(len(yy))
```

```python
fit_order1 = 1
```

```python
coeffs1 = scipy.polyfit(xx,yy,fit_order1)
```

```python
print(coeffs1) # y = ax + b
```

```python
roots1 = scipy.roots(coeffs1)
print(roots1)
```

```python
xx_new = scipy.linspace(xx[0],xx[-1],31) # get the evenly spaces x axis
print(xx_new)
```

```python
ffit1 = scipy.polyval(coeffs1,xx_new) # recompute the y
```

```python
print(ffit1)
```

```python
fit_order2 = 2
```

```python
coeffs2 = scipy.polyfit(xx,yy,fit_order2)
```

```python
print(coeffs2) # y = ax^2 + bx + c
```

```python
roots2 = scipy.roots(coeffs2)
print(roots2)
```

```python
ffit2 = scipy.polyval(coeffs2,xx_new)
```

```python
print(ffit2)
```

```python
fit_order3 = 3
coeffs3 = scipy.polyfit(xx,yy,fit_order3)
print(coeffs3) # y = ax^3 + bx^2 + cx + d
roots3 = scipy.roots(coeffs3)
print(roots3)
ffit3 = scipy.polyval(coeffs3,xx_new)
print(ffit3)
```

```python
fit_order4 = 4
coeffs4 = scipy.polyfit(xx,yy,fit_order4)
print(coeffs4) # y = ax^4 + bx^3 + cx^2 + dx + e
roots4 = scipy.roots(coeffs4)
print(roots4)
ffit4 = scipy.polyval(coeffs4,xx_new)
print(ffit4)
```

```python
p = figure(plot_width=600, plot_height=400)
p.line(xx,yy,color="red") # Original
p.circle(xx,yy,color="red")

p.line(xx,ffit1,color="blue") # fit 1st order

p.line(xx,ffit2,color="green") # fit 2nd order
p.line(xx,ffit3,color="maroon") # fit 3rd order
p.line(xx,ffit4,color="black") # fit 4th order

show(p)
```
