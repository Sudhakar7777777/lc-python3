```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Reading images
```

```python
im = np.array(Image.open('lena.jpg'))
```

```python
print(im.dtype)
# uint8

print(im.ndim)
# 3

print(im.shape)
# (512, 512, 3)

print(im.min())
print(im.max())
```

```python
plt.imshow(im)
```

```python
def rgb2gray(rgb):
    r, g, b = rgb[:,:,0], rgb[:,:,1], rgb[:,:,2]
    gray = 0.2989 * r + 0.5870 * g + 0.1140 * b
    return gray
```

```python
gray = rgb2gray(im)
```

```python
print(gray[256, 256])
```

```python
gray.shape
```

```python
gray = gray.astype(int)
print(gray[256, 256])
```

```python
plt.imshow(gray)
```

```python
print(gray[256, 256])
```

```python
th = 100
im_bool = gray > th
print(im_bool)
```

```python
im_bool_128 = (gray > th) * 255
plt.imshow(im_bool_128)
```

```python
th = 64
im_bool_64 = (gray > th) * 255
plt.imshow(im_bool_64)
```

```python
# Cropping
print(im.shape)
# (512, 512, 3)

im_trim1 = im[128:384, 128:384]
print(im_trim1.shape)
# (256, 256, 3)

plt.imshow(im_trim1)
```

```python
from scipy import ndimage
# https://docs.scipy.org/doc/scipy/reference/ndimage.html
```

```python
im_10 = ndimage.rotate(im, 10, reshape=False)
full_im_10 = ndimage.rotate(im, 10, reshape=True)
```

```python
print(type(im_10))
print(type(full_im_10))
print(im_10.shape)
print(full_im_10.shape)
```

```python
plt.imshow(im_10)
```

```python
plt.imshow(full_im_10)
```

```python
im_m10 = ndimage.rotate(im, -10, reshape=False)
full_im_m10 = ndimage.rotate(im, -10, reshape=True)
```

```python
plt.imshow(im_m10)
```
